```python
cid = pd.read_csv('worldcities.csv', sep=',')
cid['city'] = cid['city'].str.lower()
cid = cid.rename(columns={"city":"City"})
```

```python
codeList = [
        "BE", "BG", "CZ", "DK", "DE", "EE", "IE", "EL", "ES", "FR", "HR", "IT", "CY",
        "LV", "LT", "LU", "HU", "MT", "NL", "AT", "PL", "PT", "RO", "SI", "SK", "FI",
        "SE", "UK"]
cid_final = cid.loc[cid.iso2.isin(codeList)]
cid_final = cid_final.filter(items=["City", "lat","lng"])
```

```python
cid_final[cid_final["City"] == "porto"]
```

```python
datafinal2 = data_final.merge(cid_final, on="City")
datafinal2 = datafinal2.drop(columns=["Unnamed: 0"])
```
